```python
#import relevant/required modules and import csv data required for the report
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

all_liv_matches = pd.read_csv('liverpool_matches.csv')
all_chel_matches = pd.read_csv('chelsea_matches.csv')
all_man_u_matches = pd.read_csv('man_u_matches.csv')
gerrard_attributes = pd.read_csv('gerrard_attributes.csv')
lampard_attributes = pd.read_csv('lampard_attributes.csv')
```

```python
all_liv_matches.head()
```

```python
all_liv_matches.info()
```

```python
all_liv_matches['result'] = all_liv_matches['home_team_goal'] - all_liv_matches['away_team_goal']
all_liv_matches.head()
```

```python
home_games = all_liv_matches[all_liv_matches['home_team'] == "Liverpool"]
home_games.head(10)
```

```python
away_games = all_liv_matches[all_liv_matches['home_team'] != "Liverpool"]
away_games.head(10)
```

```python
h_win_loss_draw = []

for r in home_games['result']:
    if r > 0:
        h_win_loss_draw.append('W')
    elif r < 0:
        h_win_loss_draw.append('L')
    elif r == 0:
        h_win_loss_draw.append('D/T')

home_games['win_loss_draw'] = h_win_loss_draw
home_games.head()
    
```

```python
a_win_loss_draw = []

for r in away_games['result']:
    if r > 0:
        a_win_loss_draw.append('L')
    elif r < 0:
        a_win_loss_draw.append('W')
    elif r == 0:
        a_win_loss_draw.append('D/T')
        
away_games['win_loss_draw'] = a_win_loss_draw
away_games.head()
```

```python
home_games.info()
```

```python
merge_attempt = pd.concat([home_games, away_games], axis=0)
```

```python
merge_attempt
```

```python
merge_attempt.sort_values('date', axis=0)
```

```python
merge_attempt.hist();
```

```python
season_results = merge_attempt.groupby('season')['win_loss_draw'].value_counts()
```

```python
season_results.head()
```

```python
season_results2 = pd.crosstab(merge_attempt['season'], merge_attempt['win_loss_draw'])
season_results2
```

```python
season_results2.plot.bar();
```
